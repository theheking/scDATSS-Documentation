# GEPIA2 Run Through 
This script demonstrates a programmatic way to generate Kaplan-Meier survival curves using the GEPIA2 (Gene Expression Profiling Interactive Analysis) database. It specifically focuses on analyzing how the expression of certain genetic transcripts correlates with patient survival outcomes in breast cancer.

```python
import pandas as pd
import gepia
bp=gepia.survival()
bp.showParams()
```

```python
bp.setParam('dataset',"BRCA")
bp.setParam('signature',"ENST00000440973.5,ENST00000404742.5")
bp.setParam('subtype',"LuminalA")
# bp.setParam('dataset',"BRCA")
# Basal-like / Triple negative (135)

# HER2+ non-luminal (66)

# Luminal A (415)

# Luminal B (192)
from IPython.display import IFrame

result=bp.query()
IFrame(result,width=500,height=500)
```
